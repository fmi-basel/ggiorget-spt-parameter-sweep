# Compute how much tracks are lost in the filtering process

```python
def check_filter_count(df_b,concentration,dt,sweep,t,g,m):
    path = f'/tungstenfs/scratch/ggiorget/nessim/snakemake/benchmark_tracking/results/{concentration}pM/dt_{dt}/{sweep}/'
    fil = f'track_cost_cutoff_{t}_gap_closing_cost_cutoff_{g}_gap_closing_max_frame_count_{m}'
    a = pd.DataFrame()
    df_tracks_50_best = pd.DataFrame()
    counter = 0
    for file in os.listdir(path):
        if fil in file:
            df = pd.read_csv(os.path.join(path,file))
            df['seq'] = [counter]*len(df)
            df_tracks_50_best = pd.concat([df_tracks_50_best,df])
            counter +=1
    for i in range(len(df_b.seq.unique())):
        df_filter = df_b[(df_b.seq == i) & (df_b.track_cost_cutoff == t) & (df_b.gap_closing_cost_cutoff == g) & (df_b.gap_closing_max_frame_count == m)]
        if df_filter.empty:
            l = [1000]
        else:
            l = df_filter.track_id1.unique()
        f = df_tracks_50_best[df_tracks_50_best.seq == i].copy(deep=True).groupby('track_id').size()
        f = pd.DataFrame(f)
        f['in?'] = f.index.isin(l)
        a = pd.concat([a,f])
        # find the maximum length of the tracks that are not in the filtered tracks
    a[a['in?'] == False].max()
        # check if there is a track that should be filtered larger that the maximum length of the tracks that are not in the filtered tracks
    g = a[a['in?'] == True] > a[a['in?'] == False].max()
    g = g[0]
    number_of_tracks_filtered = len(g[g == True])
    number_of_tracks_filtered
    return number_of_tracks_filtered
```

```python
f = check_filter_count(df_50_10_b,50,10,'filtered_tracks',4.0,5.0,4)
```

# Build heatmap

```python
def get_best_df (df_b,df_h):
    df_b = df_b[(df_b['track_cost_cutoff']==df_h.index[0][0]) & (df_b['gap_closing_cost_cutoff']==df_h.index[0][1]) & (df_b['gap_closing_max_frame_count']==df_h.index[0][2])]
    return df_b
```

```python
df_comb_100_heat = pd.DataFrame()

for i,j,k in zip([df_100_10_h,df_100_4_h,df_100_2_h],[10,4,2],[df_100_10_b,df_100_4_b,df_100_2_b]):
    df = get_best_df(k,i)
    df['dt'] = [j]*len(df)
    df['hmean'] = [i[i.index ==i.index[0]].hmean.values[0]]*len(df)
    df['threshold'] = [i[i.index ==i.index[0]].threshold.values[0]]*len(df)
    df_comb_100_heat = pd.concat([df_comb_100_heat,df])
```

```python
df_comb_75_heat = pd.DataFrame()

for i,j,k in zip([df_75_10_h,df_75_4_h,df_75_2_h],[10,4,2],[df_75_10_b,df_75_4_b,df_75_2_b]):
    df = get_best_df(k,i)
    df['dt'] = [j]*len(df)
    df['hmean'] = [i[i.index ==i.index[0]].hmean.values[0]]*len(df)
    df['threshold'] = [i[i.index ==i.index[0]].threshold.values[0]]*len(df)
    df_comb_75_heat = pd.concat([df_comb_75_heat,df])
```

```python
df_comb_50_heat = pd.DataFrame()

for i,j,k in zip([df_50_10_h,df_50_4_h,df_50_2_h],[10,4,2],[df_50_10_b,df_50_4_b,df_50_2_b]):
    df = get_best_df(k,i)
    df['dt'] = [j]*len(df)
    df['hmean'] = [i[i.index ==i.index[0]].hmean.values[0]]*len(df)
    df['threshold'] = [i[i.index ==i.index[0]].threshold.values[0]]*len(df)
    df_comb_50_heat = pd.concat([df_comb_50_heat,df])
```

```python
df_plot = pd.DataFrame()

for i,j in zip([df_comb_50_heat,df_comb_75_heat,df_comb_100_heat],[50,75,100]):
    i['concentration'] = [j]*len(i)
    df_plot = pd.concat([df_plot,i])
```

```python
df_plot.reset_index(inplace=True)
```

```python
for i in df_plot.index:
    if df_plot.loc[i,'dt'] == 10:
        max_count = 40
    elif df_plot.loc[i,'dt'] == 4:
        max_count = 100
    else:
        max_count = 200
    
    df_plot.loc[i,'hmean'] = (3 * (df_plot.loc[i,'iou'] * (df_plot.loc[i,'count']/max_count)*(df_plot.loc[i,'most_consecutive_frames']/df_plot.loc[i,'count']))) / ((df_plot.loc[i,'iou'] *(df_plot.loc[i,'count']/max_count))+((df_plot.loc[i,'count']/max_count)*(df_plot.loc[i,'most_consecutive_frames']/df_plot.loc[i,'count']))+(df_plot.loc[i,'iou']*(df_plot.loc[i,'most_consecutive_frames']/df_plot.loc[i,'count'])))
```
